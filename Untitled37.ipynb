```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('/', '_')
)
df.columns
```

```python
critical_cols = ['row_id', 'order_id', 'customer_id']
for col in critical_cols:
    if col in df.columns:
        df = df[df[col].notna()]
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna('Unknown')

# c) For numeric columns: fill with median
num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

print("Missing values after handling:")
print(df.isna().sum())
```

```python
#5. Fix data types
# -----------------------------

# Convert dates
for date_col in ['order_date', 'ship_date']:
    if date_col in df.columns:
        df[date_col] = pd.to_datetime(df[date_col], errors='coerce')

# Convert some columns to category type
cat_like = [
    'order_priority', 'ship_mode', 'customer_segment',
    'product_category', 'product_sub-category', 'product_container',
    'country', 'region', 'state_or_province', 'city'
]
for col in cat_like:
    if col in df.columns:
        df[col] = df[col].astype('category')

# Ensure numeric columns are numeric
numeric_like = ['discount', 'unit_price', 'shipping_cost',
                'product_base_margin', 'profit', 'quantity_ordered_new', 'sales']
for col in numeric_like:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

```

```python
# -----------------------------
# 6. Feature engineering
# -----------------------------

# a) Shipping time (days)
if {'order_date', 'ship_date'}.issubset(df.columns):
    df['shipping_days'] = (df['ship_date'] - df['order_date']).dt.days

# b) Profit margin (profit as % of sales)
if {'profit', 'sales'}.issubset(df.columns):
    df['profit_margin_pct'] = np.where(
        df['sales'] != 0,
        (df['profit'] / df['sales']) * 100,
        np.nan
    )

# c) Year, month, weekday from order_date
if 'order_date' in df.columns:
    df['order_year'] = df['order_date'].dt.year
    df['order_month'] = df['order_date'].dt.month
    df['order_dayofweek'] = df['order_date'].dt.day_name()
```

```python
# -----------------------------
# 7. Basic sanity checks
# -----------------------------
print(df[['sales', 'profit', 'discount', 'unit_price']].describe())

# Example: remove negative sales if any (business rule)
df = df[df['sales'] >= 0]

# -----------------------------
# 8. Save cleaned data
# -----------------------------
clean_path = r'C:\Users\jvssu\Downloads\P6-SuperStoreUS-2015_clean.csv'
df.to_csv(clean_path, index=False)
print("Cleaned data saved to:", clean_path)
```
